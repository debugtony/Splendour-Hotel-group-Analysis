```python
import pandas as pd



# Display the first few rows of the DataFrame
print(df.head())

# Convert AvgDaily_Rate to numeric, setting errors='coerce' will turn invalid parsing into NaN
df['Avg Daily Rate'] = pd.to_numeric(df['Avg Daily Rate'].str.replace('$', ''), errors='coerce')

# Fill NaN values with a default value, for example 0
df['Avg Daily Rate'].fillna(0, inplace=True)

# Verify the changes
print(df.head())

# Optionally, save the cleaned DataFrame back to a CSV file
df.to_csv('cleaned_booking.csv', index=False)
```

```python
# Ensure the 'Revenue' column contains string values
df['Revenue'] = df['Revenue'].astype(str)

# Remove the dollar signs and commas, then convert to numeric, coercing errors to NaN
df['Revenue'] = pd.to_numeric(df['Revenue'].str.replace('$', '').str.replace(',', ''), errors='coerce')

# Fill NaN values with 0 (or another appropriate value)
df['Revenue'].fillna(0, inplace=True)
df.to_csv('cleaned_booking.csv', index=False)
```
